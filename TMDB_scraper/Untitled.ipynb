```python
import pandas as pd
```

```python
df = pd.read_csv("results.csv").drop_duplicates()
```

```python
df_1 = pd.DataFrame(df.movie_or_TV_name.value_counts()).reset_index()
df_1.columns = ['movie','count']
df_1 = df_1[:10]
df_1
```

```python
import matplotlib.pyplot as plt

plt.pie(df_1["count"], labels = df_1["movie"],wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' })
plt.show()
```
